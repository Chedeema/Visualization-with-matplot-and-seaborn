# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot')
plt.show()
```

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

```python
x = [5, 7, 8, 7, 2, 17, 2, 9]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y, color='red')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

```python
# Create multiple subplots
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 1].scatter(x, y)
axs[1, 0].bar(categories, values)
axs[1, 1].hist(data, bins=10)

plt.tight_layout()
plt.show()
```

# PROJECT

```python
data = pd.read_csv('myfaketable.csv')
data
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Add an 'Income' column with random values between 30,000 and 50,000
data['Income'] = np.random.randint(30000, 50001, size=len(data))
data
```

```python
data['Age'] = np.random.randint(20, 50, size=len(data))
data
```

```python
# Bar chart for gender distribution
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette="viridis")
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x=data['Gender'], data=data, palette= "viridis")
plt.title('Gender distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Age', y='Income', hue='Gender', palette="coolwarm", s=100)
plt.title('Age vs Income')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
location_counts = data['Location'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Location Distribution')
plt.show()
```

```python
data['Year'] = np.arange(2000, 2000 + len(data))
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Age'], color='blue', marker='o')
plt.title('Age Over Time')
plt.xlabel('Year')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=10, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Age', y='Income', data=data, scatter_kws={'s':100}, line_kws={'color':'red'})
plt.title('Age vs Income with Trendline')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data[['Age', 'Income']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Location', y='Income', hue='Gender', data=data, palette="muted")
plt.title('Income by Location and Gender')
plt.xlabel('Location')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Income', data=data, palette="Set3")
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Age', y='Income', hue='Location', palette='Set1', s=100)
plt.title('Age vs Income by Location')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D plot for Age, Income, and a random numeric variable (we'll use S/N)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(data['Age'], data['Income'], data['Year'], c='blue', marker='o')

ax.set_title('3D Plot: Age vs Income vs S/N')
ax.set_xlabel('Age')
ax.set_ylabel('Income')
ax.set_zlabel('Year')

plt.show()
```
